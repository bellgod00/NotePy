# Python 테스트 노트

```python
# 3과 5의 배수를 모두 더하기

result = 0

for n in range(1, 1000):
    if n % 3 == 0 or n % 5 == 0:
        result += n
        print(n, end=' ')

print(f"\n합계 : {result}")
```

```python

# 1부터 999까지의 숫자 중 3 또는 5의 배수의 합 구하기
result = sum(n for n in range(1, 1000) if n % 3 == 0 or n % 5 == 0)

print("합계:", result)
```

> 이런걸 작성하는 방법을 고민 생각
> 해야 하는.


```python
input1 = int(input("첫 번째 숫자를 입력하세요: "))
input2 = int(input("두 번째 숫자를 입력하세요: "))

total = input1 + input2
print("두 수의 합은 %s 입니다" % total)
```

```python
from datetime import datetime

date_str1 = '2025-12-25 00:00:00'
date_str2 = '2025-12-01 18:00:00'

date1 = datetime.strptime(date_str1, '%Y-%m-%d %H:%M:%S')
date2 = datetime.strptime(date_str2, '%Y-%m-%d %H:%M:%S') 

delta = date1 - date2
print(f"두 날짜의 차이는 {delta.days}일 입니다.")
```
